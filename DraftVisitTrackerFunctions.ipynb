```python
import pandas as pd
import requests

url = "https://raw.githubusercontent.com/ekrusins/NFLDraftSelectionTrends/refs/heads/main/WalterFootball.csv"
text = requests.get(url).text

lines = [line.strip() for line in text.split("\n") if line.strip()]
```

```python
import re

rows = []
current_season = None
current_team = None

# Regex to match player lines like:
# "Darius Alexander, Defensive Tackle, Toledo (T30)"
player_pattern = re.compile(r"^(.*?),\s*(.*?),\s*(.*?)\s*\((.*?)\)$")

for line in lines:
    # Detect season (a 4-digit year)
    if re.fullmatch(r"\d{4}", line):
        current_season = line
        continue

    # Detect team name (no commas, not a year)
    if "," not in line and not line.endswith(")"):
        current_team = line.strip()
        continue

    # Try to parse a player line
    match = player_pattern.match(line)
    if match:
        fullname = match.group(1).strip()
        position = match.group(2).strip()
        school = match.group(3).strip()
        visit_type = match.group(4).strip()

        rows.append({
            "fullname": fullname,
            "position": position,
            "school": school,
            "visiting_team": current_team,
            "season": current_season,
            "visit_type": visit_type
        })

# Convert to DataFrame
df = pd.DataFrame(rows)

# Save to CSV
df.to_csv("WalterFootballClean.csv", index=False)

df.head()
```

```python
#CBS2022

url = "https://raw.githubusercontent.com/ekrusins/NFLDraftSelectionTrends/refs/heads/main/CBSSports2022.csv"
text = requests.get(url).text

lines = [line.strip() for line in text.split("\n") if line.strip()]

rows = []
current_team = None

#Nomenclature:  "S Jaquon Brisker (Penn State)""
player_pattern = re.compile(r"^([A-Z]+)\s+(.*?)\s*\((.*?)\)$")

for line in lines:

  if "(" not in line and not line.endswith(")"):
        current_team = line.strip()
        continue

  match = player_pattern.match(line)
  if match:
        position = match.group(1).strip()
        fullname = match.group(2).strip()
        school = match.group(3).strip()

        rows.append({
            "fullname": fullname,
            "position": position,
            "school": school,
            "visiting_team": current_team,
            "season": "2022",
            "visit_type": "T30",
            "source": "CBSSports"
        })

df = pd.DataFrame(rows)
df.to_csv("CBS2022Clean.csv", index=False)
df.head()
```
